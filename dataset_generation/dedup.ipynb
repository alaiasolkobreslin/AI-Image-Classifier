```python
import pathlib
from pathlib import Path

image_dir = Path("cat_aesthetic")
```

```python
from imagededup.methods import PHash

phasher = PHash()
duplicates = phasher.find_duplicates(image_dir=image_dir, scores=True)
```

```python
duplicates
```

```python
from imagededup.utils import plot_duplicates

plot_duplicates(image_dir=image_dir, duplicate_map=duplicates, filename="000102512.png")
```

```python
duplicates_list = phasher.find_duplicates_to_remove(image_dir)
```

```python
len(duplicates_list)
```

```python
duplicates_list
```

```python
import pathlib
for fname in duplicates_list:
    path = pathlib.Path("cat_aesthetic", fname)
    path.unlink()
```
